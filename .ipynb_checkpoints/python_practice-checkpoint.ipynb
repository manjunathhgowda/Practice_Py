# 2. Swap 2 number using 3rd Variable

```python
a,b=3,6
print(f"Before swap: a={a},b={b}")
temp=a
a=b
b=temp
print(f"After swap: a={a},b={b}")
```

# 3. Swap 2 number without using 3rd Variable

```python
a,b=3,6
print(f"Before swap: a={a},b={b}")
a,b=b,a
print(f"After swap: a={a},b={b}")
```

# 10. Prime or Not

```python
n=int(input("Enter a number:"))
if n < 2:
    print("Not Prime")
else:
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
```

# 11. Factorial of a Number

```python
n=int(input("Enter a Number:"))
fact=1
for i in range(2,n+1):
    fact*=i
print(f"Factorial of Number is:{fact}")
    
```

# 12. Factorial of a number using recursion

```python
def fact(n):
    if n==0 or n==1:
        return 1
    return n*fact(n-1)
num=int(input("Enter a Number:"))
print(f"Factorial of Number is:{fact(num)}")
```

# 13. Fibonacci Series (Method-1)

```python
term=int(input("Enter the no.of term:"))
a,b=0,1
#count=0
if term<0:
    print("Not valid term")
elif term==1:
    print(f"Sequence is: {a}")
else:
    for _ in range(term):
        print(a,end=" ")
        a, b = a + b, a
```

# 14. Fibonacci Series (Method-2)

```python
term=int(input("Enter the no.of term:"))
a,b=0,1
#count=0
if term<0:
    print("Not valid term")
elif term==1:
    print(f"Sequence is: {a}")
else:
    print("Fibonacci sequence are:")
    for _ in range(term): #while count<term:
        print(a,end=" ")
        last=a+b
        a=b
        b=last
        #count+1
```

# 14. Fibonacci Series (Method-3: Recursion)

```python
def fibo(n):
    if n<=1:
        return n
    return fibo(n-1)+fibo(n-2)
term=int(input("Enter the no.of term:"))
print("Fibonacci sequence are:")
for i in range(term):
    print(fibo(i),end=" ")
```

# 15. Armstrong Number or Not (Based on String)

```python
num=input("Enter the number:")
l=len(num)
sum=0
for i in num:
    sum+=int(i)**l
if sum==int(num):
    print(f"{num} is armstrong number")
else:
    print(f"{num} is not armstrong number")
```

# 16. Armstrong Number or Not (Based on Int)

```python
num=int(input("Enter the number:"))
l=len(str(num))
sum=0
temp=num
while temp>0:
    dig=temp%10
    sum+=dig**l
    temp=temp//10
if sum==num:
    print(f"{num} is armstrong number")
else:
    print(f"{num} is not armstrong number")
```
